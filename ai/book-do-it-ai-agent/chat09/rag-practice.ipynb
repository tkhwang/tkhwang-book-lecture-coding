```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("data/OneNYC-2050-Summary.pdf")

data_nyc = loader.load()
print(data_nyc)
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)

all_splits = text_splitter.split_documents(data_nyc)

print(len(all_splits))
print(all_splits[0])
```

```python
for i, split in enumerate(all_splits):
    print(f"Split {i+1}---------------------------")
    print(split)
```

```python
loader_seoul = PyPDFLoader("./data/2040_서울도시기본계획.pdf")
data_seoul = loader_seoul.load()

text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=100)

seoul_splits = text_splitter.split_documents(data_seoul)
```

```python
for i in range(len(seoul_splits) - 1):
    seoul_splits[i].page_content += "\n" + seoul_splits[i + 1].page_content[:100]

print(seoul_splits[50].page_content)
print("---------------------------------")
print(seoul_splits[51].page_content)
```

```python
all_splits.extend(seoul_splits)
```

```python
from langchain_openai import OpenAIEmbeddings
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")

embeddings = OpenAIEmbeddings(api_key=api_key, model="text-embedding-3-large")
v = embeddings.embed_query("뉴욕의 온길가스 저감 정책은 뭐야 ?")

print(v)
print(len(v))
```

```python
from langchain_chroma import Chroma
import os

persist_directory = "./chroma_store"

if not os.path.exists(persist_directory):
    print("Creating new Chroma store")
    vectorstore = Chroma.from_documents(
        documents=seoul_splits,
        embedding=embeddings,
        persist_directory=persist_directory,
    )
else:
    print("Loading existing Chroma store")
    vectorstore = Chroma(
        persist_directory=persist_directory,
        embedding_function=embeddings
    )
```

```python
retriever = vectorstore.as_retriever(k = 3)
docs = retriever.invoke("서울시의 환경 정책이 궁금해")

for d in docs:
    print(d)
    print("---------------------------------")
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_openai import ChatOpenAI
from dotenv import load_dotenv
import os

load_dotenv()
api_key = os.getenv("OPENAI_API_KEY")

chat = ChatOpenAI(api_key=api_key)

question_answering_prompt = ChatPromptTemplate.from_messages([
    (
        "system",
        "사용자의 질문에 대해 아래 context에 기반하여 답변하라.:\n\n{context}"
    ),
    MessagesPlaceholder(variable_name="messages"),
])

document_chain = create_stuff_documents_chain(
    chat,
    question_answering_prompt
)
```

```python
from langchain.memory import ChatMessageHistory

chat_history = ChatMessageHistory()
chat_history.add_user_message("서울의 온실가스 저감 정책에 대해 알려줘.")

answer = document_chain.invoke({
    "messages": chat_history.messages,
    "context": docs
})

chat_history.add_ai_message(answer);

print(answer)
```

```python
for m in chat_history.messages:
    print(m)
```

```python
from langchain_core.output_parsers import StrOutputParser

query_for_nyc = "뉴욕은 ?"

query_augmentation_prompt = ChatPromptTemplate.from_messages([
    MessagesPlaceholder(variable_name="messages"),
    (
        "user",
        "{query}"
    ),
    (
        "system",
        """사용자가 방금 한 질문을 분석하여 완전한 질문으로 재구성하라.

        분석 과정:
        1. 이전 대화에서 다룬 주제를 파악한다
        2. 새 질문이 도시/지역 이름만 언급한다면, 이전 주제를 그 도시/지역에 적용한다
        3. 예시:
           - 이전: "서울의 온실가스 정책" → 새 질문: "뉴욕은?" → 결과: "뉴욕의 온실가스 정책은 무엇인가?"
           - 이전: "도쿄의 교통 시스템" → 새 질문: "파리는?" → 결과: "파리의 교통 시스템은 어떤가?"
        
        오직 재구성된 질문만 출력하라. 다른 설명은 하지 마라."""
    )
])

query_augmentation_chain = query_augmentation_prompt | chat | StrOutputParser()

augmented_query = query_augmentation_chain.invoke({
    "messages": chat_history.messages,
    "query": query_for_nyc
})

print(f"원래 질문: {query_for_nyc}")
print(f"확장된 질문: {augmented_query}")
```

```python
docs = retriever.invoke(augmented_query)

for doc in docs:
    print(doc)
    print("-" * 30)
```
